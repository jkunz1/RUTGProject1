```python
import pandas as pd
import requests
import json
import numpy as np
import reverse_geocoder as rg
import csv
import os
```

```python
forcl_six_path = "Resources/2016_Registered_Foreclosure_Properties.csv"
forcl_six_df = pd.read_csv(forcl_six_path)
forcl_six_df.dtypes

six=forcl_six_df.dropna(subset=["RegisteredDate"])
forcl_six_df["RegisteredDate"]= forcl_six_df["RegisteredDate"].astype('datetime64[ns]')





forcl_seven_path = "Resources/2017_Registered_Foreclosure_Properties.csv"
forcl_seven_df = pd.read_csv(forcl_seven_path)

seven=forcl_seven_df.dropna(subset=["RegisteredDate"])
forcl_seven_df["RegisteredDate"]= forcl_seven_df["RegisteredDate"].astype('datetime64[ns]')


forcl_eight_path = "Resources/2018_Registered_Foreclosure_Properties.csv"
forcl_eight_df = pd.read_csv(forcl_eight_path)

eight=forcl_eight_df.dropna(subset=["RegisteredDate"])
forcl_eight_df["RegisteredDate"]= forcl_eight_df["RegisteredDate"].astype('datetime64[ns]')
```

```python
foreclosure =pd.concat([six,seven, eight], sort=False)
foreclosure.head()
foreclosure.shape

foreclosure.to_csv(r'C:\Users\josep\Project1\Resources\merged_forclosure.csv')
```
